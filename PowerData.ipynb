```python
# seen above freq=None - there is no specific frequency in the dataset 
# the function asfreq() can "force" a frequncy onto the dataset 
#first a datetimeindex needs to be used 

times_sample = pd.to_datetime(['2013-02-03', '2013-02-06', '2013-02-08']) # arbitrarily chosen dates 
# selecting the specified dates and the 'Consumption' column 
consum_sample = opsd_daily.loc[times_sample, ['Consumption']].copy()
consum_sample
```

```python
# Converting the dato to a daily frequncy with asfreq()
conusm_freq = consum_sample.asfreq('D')
conusm_freq
```

```python
#method=ffill is forward filling the old values to avoid NaN -- CAREFUL this can also impact the data quality negatively 
conusm_freq['Consumption - Forward Fill'] = consum_sample.asfreq('D', method='ffill')
conusm_freq
```

```python
#
#
#
#
#NEW SECTION: RESAMPLING 
#
#
#
# can upsample or downsmaple 
#upsampling requires interpolation or some other data filling method 
# this section will focus on downsampling 
```

```python
#specify the data that will be included 
data_columns = ['Consumption', 'Wind', 'Solar', 'Wind+Solar']
# resampling to weekly frequency by aggregating the mean 

opsd_weekly_mean = opsd_daily[data_columns].resample('W').mean()
opsd_weekly_mean.head()

```

```python
# the interval has been reduced to 7 days as can be seen in the Date column or by "shape" comparison: 
print(opsd_daily.shape[0])
print(opsd_weekly_mean.shape[0])
```

```python
#Plotting both daily and weekly solar power in a six months period 
# defining start and end of period 

start, end = '2017-01', '2017-06'

# Plotting daily and weekly resamppled time series 
# creating the graph 
fig, ax = plt.subplots()
ax.plot(opsd_daily.loc[start:end, 'Solar'], marker='.', linestyle='-', linewidth='0.5', label='Daily Solar Power')

ax.plot(opsd_weekly_mean.loc[start:end, 'Solar'], marker='o', markersize=8, linestyle='-', label='Weekly Mean Resample')

ax.set_ylabel('Solar Production (GWh)')
ax.legend()
```

```python
#Resampling 2 
# resampling with monthly frequncy, also monthly sum instead of mean 
# also set value to NaN for months with less than 28 days of available data 
opsd_monthly = opsd_daily[data_columns].resample('M', convention = 'end').sum(min_count=28) #data_columns = ['Consumption', 'Wind', 'Solar', 'Wind+Solar']
opsd_monthly.head()
```

```python
fig, ax = plt.subplots()
ax.plot(opsd_monthly['Consumption'], color='black', label='Consumption')
opsd_monthly[['Wind', 'Solar']].plot.area(ax=ax, linewidth=0)
ax.xaxis.set_major_locator(mdates.YearLocator())
ax.legend()
ax.set_ylabel('Monthly Total (GWh)')
```

```python
# Resampling to anunal frequency 
# Computing the annual sum - also setting the value to NaN for any year which has fewer than 360 datapoints 
opsd_annual = opsd_daily[data_columns].resample('A').sum(min_count=360)
# The default index of the resampled DataFrame is the last day of each year,
# ('2006-12-31', '2007-12-31', etc.) so to make life easier, set the index
# to the year component
opsd_annual = opsd_annual.set_index(opsd_annual.index.year)
opsd_annual.index.name = 'Year'
# computing the ratio of Wind+Solar to consumption 
opsd_annual['Wind+Solar Ratio in %'] = (opsd_annual['Wind+Solar'] / opsd_annual['Consumption'])*100
opsd_annual.tail()
```

```python
# Plotting from 2012 onwards, because there is no solar production in earlier years
ax = opsd_annual.loc[2011:, 'Wind+Solar Ratio in %'].plot.bar(color='C0')
ax.set_ylabel('Fraction')
ax.set_ylim(0,0.3)
ax.set_title('Wind + Solar Share of Annual Electricity Consumption')
plt.xticks(rotation = 0)
```
